```python
sp500_tickers = pd.read_csv("../data/S&P500-Symbols.csv")
```

### Download S&P 500 Stock History

```python
sp500_tickers_str = ' '.join(list(sp500_tickers['Symbol']))
```

```python
# last year of prices
raw_data = yf.download(sp500_tickers_str, start="2018-01-02", end="2020-12-11")
```

```python
raw_data.head()
```

```python
data = raw_data.stack(1).reset_index().rename(columns = {'level_1':'Ticker'}).set_index('Date')
```

```python
data.head()
```

```python
data.to_csv("../data/sp500_prices_2018-2020.csv")
```
