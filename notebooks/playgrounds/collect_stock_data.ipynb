# Collect and Store Stock Data

```python
#from get_all_tickers import get_tickers as gt
import pandas as pd
import yfinance as yf
```

## S&P 500 Tickers

```python
# temporary solution is reading the tickers from wikipedia
table=pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
df = table[0]
df.to_csv('../../data/S&P500-Info.csv')
df.to_csv("../../data/S&P500-Symbols.csv", columns=['Symbol'], index = False)
```

```python
df
```

```python
sp500_tickers = pd.read_csv("../../data/S&P500-Symbols.csv")
```

## Yfinance: Download S&P 500 Stock History

```python
sp500_tickers_str = ' '.join(list(sp500_tickers['Symbol']))
```

```python
# last year of prices
recent_data = yf.download(sp500_tickers_str, start="2018-01-02", end="2020-12-11")
```

```python
recent_data.head()
```

```python
recent_data = recent_data.stack(1).reset_index().rename(columns = {'level_1':'Ticker'}).set_index('Date')
```

```python
recent_data.head()
```

### Basic csv storage

```python
recent_data.to_csv("../../data/sp500_prices_2018-2020.csv")
```

```python
recent_data[recent_data['Ticker'] == 'EBAY'].to_csv("../data/ebay_prices_2018-2021.csv")
```

## Storing Data Efficiently Using HDF5

```python
historical_data = yf.download(sp500_tickers_str)
```

```python
historical_data = historical_data.stack(1).reset_index().rename(columns = {'level_1':'Ticker'}).set_index('Date')
```

```python
historical_data.head()
```

```python
historical_data.info()
```

### Writing

```python
h5 = pd.HDFStore('../../data/sp500_history.h5', 'w')
```

```python
%time h5['data'] = historical_data
```

```python
h5
```

```python
h5.close()
```

### Reading

```python
h5 = pd.HDFStore('../../data/sp500_history.h5', 'r')
```

```python
%time historical_data_read = h5['data']
```

```python
historical_data_read.info()
```

```python
h5.close()
```

### More Flexible Write to HDF Store

```python
historical_data.to_hdf('../../data/sp500_flex.h5', key = 'historical_data', format = 'table')
sp500_tickers.to_hdf('../../data/sp500_flex.h5', key = 'tickers', format = 'table')
```

```python
pd.read_hdf('../../data/sp500_flex.h5', key = 'tickers')
```

```python
%time pd.read_hdf('../../data/sp500_flex.h5', key = 'historical_data')
```
