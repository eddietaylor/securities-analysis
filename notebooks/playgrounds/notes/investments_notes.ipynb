## 1.1 Real Assets vs. Financial Assets

-Wealth of a society is determined by the productive capacity of its economy aka the goods and services its member can create!\
-The capacity is a function of the **real assets** of the economy: the land, buildings, machines, and knowledge that can be used to produce goods and services\
-**Financial assets** do not contribute to the productive capacity of a society but they are claims on the real assets that do. They are claims on the income generated from these real assets!!! \
-Buy shares and share in the income generated from using said real assets! \
-Firms use this money to buy more real assets \
-When aggregating all of the balance sheets in the economy, the financial assets and liabilities cancel each other and the net wealth of the economy is all real assets! \
-Success or failure of financial assets ultimately depends on performance of underlying real assets!!!

## 1.2 Financial Assets

-3 broad type of financial assets: \
1.) fixed income \
2.) equity \
3.) derivatives

-**fixed income** (debt securities) promises fixed stream of income. e.g. corporate bond promises fixed interest. So least tied to performance of issuer (unless they go bankrupt!)

-**Money Market** refers to debt securities that are short-term, highly marketable, and low-risk \
  * e.g. Treasury bills or Bank Certificate of Deposit (CDs) 
  
-fixed income **Captital Market**  includes long-term securities \
  * e.g. Treasury bonds as well as bonds issued by federal agencies, corporations, municipalities

-**equity** represent ownership share in the corporation. Not promised any payment. Directly tied to success of the firm and it real assets and therefore its riskier

-**derivative securities** such as options and futures contracts provide payoffs determined by prices of other assets such as bonds or stocks. Name because their value derives from value of other financial assets. \
One use of derivatives is to hedge risks, aka transfer them to other parties

## 1.4 The Investment Process

-The portfolio is simply the collection of investment assets. 

-These assets can be categorized into asset classes: stocks, bonds, real estate, commodities, etc.

-In constructing a portfolio you decide on the **asset allocation** (choice of the asset classes) and the **security selection** (within each asset class, which securities to hold)

-Top-down starts with asset allocation

-bottom-up constructs with securities that seem attractively priced without care for asset allocation

## 1.5 Markets Are Competitive

-highly competitive markets means no free lunches! Implications:

  * The Risk-Return Trade-Off

  * Efficient Markets

No free lunches means we should rarely find bargains. Markets process all available information about securities quickly and efficiently and this is reflected in the price. If it were true, there would neither be underprice nor overpriced securities. 

**Passive management** refers to diversified portfolios without attempting to improve performance through security analysis

**Active managment** attempts to improve performance by identifying mispriced securities or by timing performance of broad asset classes

If markets are efficient, it's probably better to do passive management. But if it were true, there would be no point in doing security analysis.

## 2.1 The Money Market

-short-term debt securities, highly marketable, trade in large portions

-e.g. T-bills, money market mutual funds (accessible to small investors)

-T-bills

  * one way government raises money
  * buyer buys at discount and at maturity is paid the face value (par value)
  * maturities range from 4 weeks to 52 weeks
  * sell at denominations of $\$$100 but $10000 more common
  * Income earned from them exempt from state and local taxes!
  
  * Financial press reports **yields** for both **bid price** and **ask price**
  * ask price is what you would pay to buy a T-bill from securities dealer
  * bid price is the slightly lower price that you would sell to dealer
  * The **bid-ask spread** is the difference and it is the dealer's source of profit
  * The bid price is like the maximum willingness to pay for the dealer
  * The ask price is like the minimum willingness to sell for the dealer
  * When the bid-ask spread is narrow, it means the security is very liquid
  * Inverse relationship between the price and the yield (see https://xtbs.com.au/insights/price-yield-inverse-relationship/)
  * Also see Fidelity for a great explanation on relationship between interest rates, treasure prices, and yields (https://www.fidelity.com/learning-center/investment-products/fixed-income-bonds/bond-prices-rates-yields)
  * quoted as an annualized discount from par! (see below figure)

### Certificates of Deposit

-time deposit with the bank

-may not be withdrawn and the banks pays you the principal with interest at the end of the fixed term

### Commercial Paper

-short term debt notes issued by companies

-dollar denominated deposits at foreign banks so banks can escape regulation

### Repos and Reverses

-government security sold on overnight basis (the security is collateral)

### Federal Funds

-banks maintain their own deposits in federal reserve

  * fed funds rate key barometer of monetary policy

### Brokers' Calls

-Buying stocks on margin means borrowing money from broker

-Broker then borrows from bank, promising to pay back immediately (on call)

-The rates on these are usually 1% higher than T-bills

### The LIBOR Market

## 2.2 The Bond Market

-longer term borrowing which includes Treasure notes and bonds, corporate bonds, municipal bonds, mortgage securities

### Treasure Notes and Bonds

-longer term (10 - 30 years)

-offer coupon payments (semiannual interest payments)

-prices quoted as percentage of par ($1000). See below figure.

-yield to maturity is the yield if all the coupons are reinvested

-**Yield** looks at change in interest or dividends and does not include capital gains (increase in security price) 

  * Obviously yield can only be applied to a narrow range of securities
  * return includes interest, dividends, and capital gains
  * There several different types of yield: **coupon rate**, **nominal yield**, **current yield**, and **yield-to-maturity**
  * **Current yield** is an investment's annual income (interest and dividends) divided by its current price (which fluctuates)
  * **Nominal yield** is an investment's annual income (interest and dividends) divided by its par value
  * The **Coupon rate** is the **Nominal yield** and it is fixed!
  * Investors who pay a premium for a bond above its par value, will earn a rate of return thats less than nominal yield
  * Nominal yield (aka coupon rate) depends on inflation, real interest rate, and credit risk of issuer
  * Nominal yield is the rate of inflation plus the real interest rate (credit risky issuers also have higher yields)

## 2.3 Equity Securities

### Common Stock as Ownership Shares

-*Common stock* aka *equity securities* aka *equities* are ownership shares in a corporation

  * Common because owners don't own any particular part of assets but rather the assets are common to all
  * each share entitles you to 1 vote on any matters of corporate governance that are put to the vote at annual meeting
  * also entitles you to share of any financial benefits
  * corporation is controlled by board of directors elected by shareholders
  * the board selects managers to run the corporation and makes sure they are in line with shareholder's interests
  * managers usually control annual vote by proxy vote (people who don't attend give their votes away)

### Characteristics of Common Stock

-**Residual Claim**
  * Shareholders have final claims on assets and income. If there is a liquidation, tax authorities, employees, suppliers, bondholders, and other creditors are paid first
  * In normal times, shareholders have final claims on operating income after interest and tax is paid. 
  * Management can pay out this residual through dividends or reinvest and make the share price go up
  
-**Limited Liability**
  * creditors can't come after shareholders' personal assets. For unincorporated businesses, creditors can come after owners' personal assets

### Stock Market Listings

-Listings include Close, Dividend (paid quarterly), Yield (annual dividend per dollar paid for the stock...remember yield does not include capital gains)

-**Capital Gains** are part of the return generated from price increases or decreases

  * low dividend firms offer greater potential for capital gains otherwise who would buy?
  
-**Price-Earnings Ratio** is the raio of the current stock price to last year's earnings per share

  * how much you have to pay per dollar of earnings
  
-Where yield and P/E ratio not reported, company doesn't distribute dividends and has zero or negative earnings

### Preferred Stock

-Like debt security in the sense that its fixed income and you don't get voting rights
-Like an equity security in that firm retains discretion to pay the dividend
-Lower yields than bonds

### Dow Jones Industrial Average

-30 large "blue-chip" corporations index computed since 1896
-Used to be a simple average of the prices
-Because of this summing of prices, it is **price-weighted**
-But the divisor is no longer 30
  * this divisor is adjusted for stock splits, dividend payments of more than 10%, or when one company in the 30 is replaced by another so as to keep the index continuous
  
  $$\text{DJIA} = {\sum p \over d}$$
  
  $$\text{DJIA} = {\sum p_\text{old} \over d_\text{old} } = {\sum p_\text{new} \over d_\text{new} }$$
