# The Investment Environment

-Investment: Current commitment of money or other resources in the expectation of reaping future benefits

-When buying stock, hopefully future benefits justify time money is tied up and the risk

   * sacrificing time that could be spent elsewhere

-In general, investing involves sacrificing something of value now, expecting to benefit later

  * discuss relationship between real assets (produce the goods and services) and securities/financial assets 
  * why financial assets are important for a functioning economy

-Central themes to investing:

  * Risk-Return Tradeoff

## 1.1 Real Assets vs. Financial Assets

-Wealth of a society is determined by the productive capacity of its economy aka the goods and services its member can create!

-The capacity is a function of the **real assets** of the economy: the land, buildings, machines, and knowledge that can be used to produce goods and services

-**Financial assets** do not contribute to the productive capacity of a society but they are claims on the real assets that do. They are claims on the income generated from these real assets!!!

-Buy shares and share in the income generated from using said real assets!

-Firms use this money to buy more real assets

-When aggregating all of the balance sheets in the economy, the financial assets and liabilities cancel each other and the net wealth of the economy is all real assets!

-Success or failure of financial assets ultimately depends on performance of underlying real assets!!!

## 1.2 Financial Assets

-3 broad types of financial assets:

1.) fixed income

2.) equity

3.) derivatives

-**Fixed Income** (debt securities) promises fixed stream of income. e.g. corporate bond promises fixed interest. So least tied to performance of issuer (unless they go bankrupt!)

   * Money Market refers to debt securities that are short-term, highly marketable, and low-risk

     * e.g. Treasury bills or Bank Certificate of Deposit (CDs) 
  
   * fixed income capital market includes long-term securities

     * e.g. Treasury bonds as well as bonds issued by federal agencies, corporations, municipalities

-**Equity** represent ownership share in the corporation. Not promised any payment. Directly tied to success of the firm and its real assets and therefore its riskier

-**Derivative Securities** such as options and futures contracts provide payoffs determined by prices of other assets such as bonds or stocks. Named so because their value derives from value of other financial assets.
One use of derivatives is to hedge risks, aka transfer them to other parties

## 1.4 The Investment Process

-The portfolio is simply the collection of investment assets. 

-These assets can be categorized into asset classes: stocks, bonds, real estate, commodities, etc.

-In constructing a portfolio you decide on the **asset allocation** (choice of the asset classes) and the **security selection** (within each asset class, which securities to hold)

-Top-down starts with asset allocation

-bottom-up constructs with securities that seem attractively priced without care for asset allocation

## 1.5 Markets Are Competitive

-highly competitive markets means no free lunches! Implications:

  * The Risk-Return Trade-Off

  * Efficient Markets

No free lunches means we should rarely find bargains. Markets process all available information about securities quickly and efficiently and this is reflected in the price. If it were true, there would neither be underpriced nor overpriced securities. 

**Passive management** refers to diversified portfolios without attempting to improve performance through security analysis

**Active managment** attempts to improve performance by identifying mispriced securities or by timing performance of broad asset classes

If markets are efficient, it's probably better to do passive management. But if it were true, there would be no point in doing security analysis.

## 2.1 The Money Market

-short-term debt securities, highly marketable, trade in large portions

-e.g. T-bills, money market mutual funds (accessible to small investors)

-T-bills

  * one way government raises money
  * buyer buys at discount and at maturity is paid the face value (par value)
  * maturities range from 4 weeks to 52 weeks
  * sell at denominations of $\$$100 but $10000 more common
  * Income earned from them exempt from state and local taxes!
  
  * Financial press reports **yields** for both **bid price** and **ask price**
  * ask price is what you would pay to buy a T-bill from securities dealer
  * bid price is the slightly lower price that you would sell to dealer
  * The **bid-ask spread** is the difference and it is the dealer's source of profit
  * The bid price is like the maximum willingness to pay for the dealer
  * The ask price is like the minimum willingness to sell for the dealer
  * When the bid-ask spread is narrow, it means the security is very liquid
  * Inverse relationship between the price and the yield (see https://xtbs.com.au/insights/price-yield-inverse-relationship/)
  * Also see Fidelity for a great explanation on relationship between interest rates, treasure prices, and yields (https://www.fidelity.com/learning-center/investment-products/fixed-income-bonds/bond-prices-rates-yields)
  * quoted as an annualized discount from par! (see below figure)

## 2.2 The Bond Market

-longer term borrowing which includes Treasury notes and bonds, corporate bonds, municipal bonds, mortgage securities

### Treasure Notes and Bonds

-longer term (10 - 30 years)

-offer coupon payments (semiannual interest payments)

-prices quoted as percentage of par ($1000). See below figure.

-yield to maturity is the yield if all the coupons are reinvested

-**Yield** looks at change in interest or dividends and does not include capital gains (increase in security price) 

  * Obviously yield can only be applied to a narrow range of securities
  * return includes interest, dividends, and capital gains
  * There several different types of yield: **coupon rate**, **nominal yield**, **current yield**, and **yield-to-maturity**
  * **Current yield** is an investment's annual income (interest and dividends; in the case of bonds its interest) divided by its current price (which fluctuates)
  * **Nominal yield** is an investment's annual income (interest and dividends) divided by its par value
  * The **Coupon rate** is the **Nominal yield** and it is fixed!
  * Investors who pay a premium for a bond above its par value, will earn a rate of return thats less than nominal yield
  * Nominal yield (aka coupon rate) depends on inflation, real interest rate, and credit risk of issuer
  * Nominal yield is the rate of inflation plus the real interest rate (credit risky issuers also have higher yields)

## 2.3 Equity Securities

### Common Stock as Ownership Shares

-*Common stock* aka *equity securities* aka *equities* are ownership shares in a corporation

  * Common because owners don't own any particular part of assets but rather the assets are common to all
  * each share entitles you to 1 vote on any matters of corporate governance that are put to the vote at annual meeting
  * also entitles you to share of any financial benefits
  * corporation is controlled by board of directors elected by shareholders
  * the board selects managers to run the corporation and makes sure they are in line with shareholder's interests
  * managers usually control annual vote by proxy vote (people who don't attend give their votes away)

### Characteristics of Common Stock

-**Residual Claim**
  * Shareholders have final claims on assets and income. If there is a liquidation, tax authorities, employees, suppliers, bondholders, and other creditors are paid first
  * In normal times, shareholders have final claims on operating income after interest and tax is paid. 
  * Management can pay out this residual through dividends or reinvest and make the share price go up
  
-**Limited Liability**
  * creditors can't come after shareholders' personal assets. For unincorporated businesses, creditors can come after owners' personal assets

### Stock Market Listings

-Listings include Close, Dividend (paid quarterly), Yield (annual dividend per dollar paid for the stock...remember yield does not include capital gains)

-**Capital Gains** are part of the return generated from price increases or decreases

  * low dividend firms offer greater potential for capital gains otherwise who would buy?
  
-**Price-Earnings Ratio** is the ratio of the current stock price to last year's earnings per share

  * how much you have to pay per dollar of earnings
  
-Where yield and P/E ratio not reported, company doesn't distribute dividends and has zero or negative earnings

### Preferred Stock

-Like debt security in the sense that its fixed income and you don't get voting rights

-Like an equity security in that firm retains discretion to pay the dividend

-Lower yields than bonds

## 2.4 Stock and Bond Market Indexes

### Stock Market Indexes

-DJI, other broad indexes and foreign indexes such as Nikkei Average of Tokyo and Financial Times index of London published daily

### Dow Jones Industrial Average

-30 large "blue-chip" corporations index computed since 1896

-Used to be a simple average of the prices

-Because of this summing of prices, it is **price-weighted**

-But the divisor is no longer 30

  * this divisor is adjusted for stock splits, dividend payments of more than 10%, or when one company in the 30 is replaced by another so as to keep the index continuous
  
  $$\text{DJIA} = {\sum p \over d}$$
  
  $$\text{DJIA} = {\sum p_\text{old} \over d_\text{old} } = {\sum p_\text{new} \over d_\text{new} }$$
  
  where p are the prices of the component stocks and d is the Dow Divisor
  
-Tries to be representative of the broad market

### Standard & Poor's 500 Index

-Better than DJIA because its 500 companies and its market cap weighted or **value-weighted**

-Essentially an average of the market capitalization of these 500 companies (market cap of publicly traded stocks! (free float))

-The divisor is once again adjusted to keep the S&P 500 continuous (but stock splits do not affect the market caps!)
  * additional share issuance, stock buyback, special dividends, etc. affects the divisor
  
  $$ \text{Index Level} = {\sum \left({P_i} \cdot {Q_i}\right) \over Divisor}$$
  
-Beauty of S&P 500 is that it reflects a straightforward portfolio strategy, namely, buy shares of the S&P 500 in proportion to their market capitalization. The index tracks the capital gains of this portfolio (see below calculation and plot)

-can invest through index funds or ETFs

### Options

-**Call option** gives holder right to *purchase* an asset at the **exercise** or **strike price** at any time up to and including the expiration date

-**Put option** gives holder right to *sell* as asset for a specified exercise price on or before expiration date. 

    * In this case, profit will be had when asset price decreases

-Futures contracts call for delivery of an asset at a delivery date for a specified price, the futures price.

## 3.2 How Securities Are Traded

-Financial markets develop to meet the needs of particular traders

-imagine somebody wishing to invest in a financial asset would have to find someone willing to sell

   * eventually these people would find each other and meetups at a certain venue (this is how modern markets were formed)

### Types of Markets

4 types of markets: direct search, brokered, dealer, and auction

### Direct Search Markets

-least organized and buyers and sellers find each other through direct search (craigslist, facebook marketplace)

-participation is sporadic

### Brokered Markets

-next level of organization

-where participation is more active, people will pay for search services (broker)

   * this happens in real estate
   * brokers also develop knowledge in valuing assets traded in the market
   * primary market is also an example of a brokered market where IB is the one finding a buyer

### Dealer Markets

-when activity gets to the next level (number of transactions I suppose) for a particular type of asset, dealers emerge

-dealers buy the asset and then later sell them for profit 

   * the bid-ask spread is their profit
    
   * they save traders search costs because they can easily look up dealer's prices
    
   * there needs to be a lot of market activity before dealing is an attractive source of income

### Auction Markets

-most integrated market

-traders converge on one location either physically or electronically

   * NYSE is an example of an auction market
   * one doesn't need to search across dealers to find the best price
   * participants can save bid-ask spread (really? not my experience)

### Types of Orders

2 types of orders: market orders and orders contingent on price

### Market Orders

-market orders are buy or sell orders to be executed immediately at the current market prices

-you're quoted a bid and ask price

   * you'll pay the ask price (thats the minimum amount they're willing to sell you)
   * and if you sell it back, you'll get it at the bid price
   * These bid and ask prices are good only up to a certain amount of shares (**depth**)
   * After you exceed the depth, you might get charged more for the remaining than what you were quoted!
   * **depth** is a component of liquidity (its larger in S&P 500 than Russell 2000 (smaller stocks))
   * Another trader can beat you to the quote!
   * Even more, the quote can change before the order arrives, causing execution at a different price than the one quoted

### Price-Contingent Orders

-limit buy order specifies maximum to buy stock at

-limit sell order specifies minimum to sell stock at

-collection of **limit orders** is a *limit order book* (see below)

   * best orders at the top (highest limit buy order on top then descends, lowest limit sell order on top then ascends) 
   * best limit orders are the *inside quotes*
   * the *inside quotes* have small order sizes however
   * an investor looking to purchase larger order size will face an *effectice* bid-ask spread

## 3.5 New Trading Strategies

-These technologies have led to the rise of algorithmic trading and high frequency trading

### Algorithmic Trading

-use of computer programs to make trading decisions

-half of equity volume in US believed to be initiated by computer

-utilize cross-market price comparison!

-Some strategies include:

    * exploiting very short-term trends (seconds) as new information about a firm or sentiment becomes available
    * pairs trading where normal relations between pairs or groups of stocks seem temporarily disrupted and a small profit can be had as they move back into alignment
    * exploting discrepancies between stock prices and prices of stock-index futures contracts
    * others attempt to profit from bid-ask spread

### High-Frequency Trading

-high frequency trades on the order of milliseconds and even microseconds now

-being the first to execute trade so they "co-locate" next to stock exchange

-they don't want to have their actions influence the price or to have HFTs infer their strategies

## 3.8 Buying on Margin

-investors have access to a form of debt financing called *broker's call loans* and using this is called *buying on margin*

-buying on margin means the trader purchases a portion (the margin) of the price and borrows the rest

    * brokers borrow money from banks at a certain rate, and charge investor this rate PLUS a service charge
    
    *regulations require the margin to be at least 50%
    
-percentage margin is the ratio of equity to the market value of the stock in the account

  $$ 1 + r_{real} = \frac{1 + r_{nom}}{1 + i}$$
  
  $$ r_{real} \approx r_{nom} - i$$

## 5.4 Risk and Risk Premiums

### Holding Period Return

-Essentially the return you get during a certain investment horizon

$$ \text{HPR} = \frac{\text{Ending price of a share - Beginning price + Cash dividend}}{\text{Beginning price}} $$

-Essentially, the return on equity is the sum of the **dividend yield** and **capital gains** 

-percentage return from dividend is the **dividend yield**

### Expected Return and Standard Deviation

-if you have an investment with a certain horizon and probability distribution of HPR at that time, you can calculate the **expected return**
  * Just the expectation value of the return over this pdf of returns at the end of the time horizon
  * Can also calculate the volatility and standard deviation with this probability distribution and expectation value

### Excess Returns and Risk Premiums

-How much would you invest in a financial asset?

  * Depends on the expected reward for the given risk
  
-Measure the reward as the difference in the *expected* HPR and the **risk-free rate** (rate earned from risk-free assets such as T-bills, money market funds, or the bank

  * We call this the **risk premium**
  
  $$ \text{Risk Premium} = E[r] - r_{f} = E[r - r_{f}] $$
  
  
  * The difference betrween the *actual* return and the risk-free rate is the **excess return**
  
  $$ \text{Excess Return} = r - r_{f} $$
  
  * The risk is the standard deviation of the excess return
  
  $$ \text{Risk} = \sigma[r - r_{f}] = \sqrt{var[(r - r_f)]} $$
  
-**risk averse** investors require a risk premium to invest

-Note: The benchmark risk-free rate (should be real) should have a maturity which matches the investment horizon! 


## 5.5 Time Series Analysis of Past Rates of Return

### Time Series vs. Scenario Analysis

-In forward-looking scenario analysis (above), we determine a set of scenarios with associated HPRs and probabilities and calculate the risk premium (reward) and standard deviation (risk)

-In contrast, asset return histories do not come with the probabilities at the time of observation and we must infer the probability distributions 

### Expected Returns and the Arithmetic Average

-Equally weight each observation and simply do an average to calculate the expected return

  * arithmetic average equivalent to summing over a uniform probability distribution

### The Geometric (Time-weighted) Average Return

-arithmetic tells us about the *expected* return but what about the *actual* performance? 

-This is essentially the EAR (you're uncompounding to get the annual)

-In general, the geometric average of returns is less than the arithmetic average of returns, and if the returns are normal, the difference is:

$$ E[\text{Geometric average}] = E[\text{Arithmetic average}] - \frac{1}{2}\sigma^{2} $$

### Variance and Standard Deviation

-With regards to risk, we are interested in the deviations of actual outcomes from the *expected* return

  * but we cannot measure the expected return so we estimate it with the arithmetic average
  
  $$ \hat{\sigma}^2 = \frac{1}{n} \sum_{s=1}^{n} [r(s)-\bar{r}] $$
  
-However, this estimate is biased downward, because we have taken deviations from the sample arithmetic average instead of the unknown expectation value of the returns

  * To fix, simply multiply by (becomes relevant for large sample sizes) $$ \frac{n}{n-1} $$

### ***Mean and Standard Deviation Estimates from Higher-Frequency Observations

-frequency of observations does not improve estimate of expected return!!!

  * intra-year observations, for example, provide no additional information about mean return, since they have to be consistent with the yearly return

-duration of observations does improve estimate accuracy of estimate of expected return

  * This suggests a rule: Use the longest sample that you believe comes from the same return distribution!!

-But the accuracy of estimates of the standard deviation can be increased with increased number of observations

  * more frequent observations gives us more information about the distribution of deviations from the average
  
-Estimates of the standard deviation begin with the variance. *When the returns are uncorrelated*, the variances add up

$$ \sigma_A = \sqrt{12}\sigma_M $$

### The Reward-to-Volatility (Sharpe) Ratio

-Essentially take the ratio of the reward (risk premium) to the risk (standard deviation of the excess return)

$$ \text{Sharpe Ratio} = \frac{E[R_a - R_f]}{\sqrt{var[R_a - R_f]}} $$

-Used to evaluate the performance of investment managers

-Adequate measure of the risk-return trade-off for diversified portfolios but inadequate for individual assets such as shares of stock!!!

## 5.7 Deviations From Normality and Alternative Risk Measures

-deviations from normality measured by greater than second order moments of the distribution

-skewness, the third moment, is the expectation of the cubed deviations from the mean over the cubed standard deviation

$$ \text{Skew} = E[\frac{(R-\bar{R})^3}{\hat{\sigma}^3}] $$

-When the returns are positively skewed, rare extreme positive events may occur and the distribution is "skewed" to the right

  * The standard deviation tends to overestimate the risk in this case! (Just like how the mean will overestimate the 50th percentile with positive skewness)
  * Likewise the standard deviation underestimates the risk with negatively skewed distributions
  
-kurtosis, the fourth moment, the expectation of the deviations from the mean over the standard deviation to the fourth power

$$ \text{Kurt} = E[\frac{(R-\bar{R})^4}{\hat{\sigma}^4}] - 3 $$

-This concerns the increased likelihood of extreme events on either side in the form of "fat tails"

-With high kurtosis, the standard deviation would underestimate the likelihood of extreme events

-Apparently skewness and kurtosis do not changes when annualized from high frequency observations!!!

### Value at Risk

-simply the loss corresponding to a very low percentile of the returns distribution

-Typically its the 1% quantile of the distribution (99% of the returns will exceed it)

### Expected Shortfall

-VaR is optimistic because it takes the highest values of that those returns lying below the 1% quantile

-The **expected shortfall** or **conditional tail expectation** is simply the expectation value of that worst 1% of returns and is more informative

### Lower Partial Standard Deviation and the Sortino Ratio

-issue with using standard deviation for risk is asymmetry in distributions and the fact that we should really be looking at deviations from the risk free rate and not the mean rate of return

-**Lower partial standard deviation (LPSD)** is essentially the standard deviation of the deviations from the risk-free rate, conditioned on it being a negative deviation

-When you replace the denominator of the Sharpe Ratio with LPSD, it is the **Sortino Ratio**

### Relative Frequency of Large, Negative 3-Sigma Returns

-Essentially compare the frequency of negative 3 sigma events to frequency of such events for a normal

## 5.9 Normality and Long-Term Investments

****Even if short-term stock returns were normal, long term would not be!***
  * This is due to the fact that the returns do not add through time, they multiply (aka compound) 
  
  $$ (1 + r_1)(1 + r_2) \text{  not  } r_1 + r_2 $$
  
-In the long-run, the returns distribution begins to look lognormal with a positive skewness
  * the upside is unlimited but downside has a floor (you can't lose more than you invested
  
  * but using continuous compounding, the compounded returns are multiplied exponentials and the exponent then is additive in the returns so it stays normal (just have to log transform the returns)
  
  $$ \exp{r_{cc}(1)}*\exp{r_{cc}(2)} = \exp{r_{cc}(1) + r_{cc}(2)} $$
  
  * In addition, if the returns are the same over each holding period, then you just do n number of holding periods times the return
  * Also, if the returns are uncorrelated, the variance is also just n times the variance of the single holding period. 
  * generalization to arbitrary investment horizons is:
  
  $$ E(W_{T}) = W_{0}*\exp{E(r_{cc})T} $$
  
  Where $$ E(W_{T}) $$ is the expected final value of the portfolio
  
  Also, the variance is proportional to the time horizon:
  
  $$ Var(r_{cc}T) = TVar(r_{cc}) $$
  
  Therefore,
  
  $$ \sigma(r_{cc}T) = \sqrt{T}\sigma $$

### Short-Run versus Long-Run Risk

-This analysis appears to offer a relief from investment risk: **Because expected return increases faster with horizon than risk (std increases as sqrt of T while the expected return increases with T), you can argue that investments are less risky when horizons are large**

  * but worse case scenarios get worse, even though less likely
  * The 1% VaR grows with investment horizon so the *probability* of losses falls with investment horizon but the *magnitude* of the potential losses grows
  
-better way of quantifying risk of long-term investment is to calculate the market price of insuring it against a loss

### Risk Aversion and Utility Values

-risk tends to increase with increased risk premium... so how do you score? of course penalize higher risk and reward higher return

-Utility refers to how much benefit investors derive from portfolio performance

-common utility function is as follows:

$$ U = E(r) - \frac{1}{2}A\sigma^2 $$

Where A is the investor's degree of risk aversion (how much their utility decreases with increasing risk)

-Investors select portfolio with highest utility! 

-**Certainty Equivalent Rate** is the rate that a risk-free investment would need to have to have the same utility as the portfolio
  * The investor would only choose the portfolio if its certainty equivalent rate is greater than the risk free rate
  * Essentially, it is the guaranteed return that the investor would accept now (since its utility is equal to or greater than the proposed portfolio)

-**Risk-neutral** investors (A=0) only judge a portfolio off of its expected returns

-**Risk lover** investors (A<0) happily engage in fair games and gambles

-One can plot expected return vs the risk. The northwest the direction is the preferred one since this increases expected return and reduces risk (for a risk averse investor)

-**mean-variance (M-V) criterion**
  * Portfolio A dominates B if:
  $$ E(r_A) >= E(r_B) $$
  and
  $$ \sigma_A <= \sigma_B $$

-Can find the **indifference curve** by setting the utility to a constant and plotting all of the expected return and standard deviation values which satisfy this

-You can take surveys which quantify your risk aversion!

## 6.4 Portfolios of One Risky Asset and a Risk-Free Asset

-Keep in mind the difference between asset allocation and capital allocation
  * *Asset Allocation* is the adjustment of the weights of the risky part of the portfolio (the allotment of funds between stocks and funds)
  * *Capital Allocation* is the adjustment of the weights between the risk-free investments and the risky assets

-let y by the proportion of funds allocated to the risky asset and therefore (1-y) be allocated to tghe risk-free asset. If the return on the risky asset is r_p and the return on the risk-free asset is r_f then the return on the complete portfolio is:

$$ r_c = yr_p + (1-y)r_f $$

And the expected return is:

$$ E(r_c) = yE(r_p) + (1-y)r_f $$

Rearranging, we arrive at an intuitive formula:

$$ E(r_c) = r_f + y(E(r_p) - r_f) $$

Essentially the base rate of return is the risk-free rate and the additional return is the proportion of the risky asset times the risk premium

Also, the standard deviation of the portfolio is essentially the std of the risky asset times its proportion (from stat, if you muiltiply a r.v. by a constant, the std is multiplied by that constant)

$$ \sigma_c = y\sigma_p $$

Using this, one can solve for the expected return of the portfolio as a function of the standard deviation

$$ E(r_c) = r_f + \sigma_c(\frac{E(r_p) - r_f}{\sigma_p}) $$

Notice this is a line in the expected rate of return - standard deviation plane and the slope is the Sharpe Ratio. This line is what is referred to as the Capital Allocation Line (CAL)

-Essentially, this line is parametrized by y the proportion allocated to the risky asset. Increasing y represents moving northeast on this line.

-The above represents the **investment opportunity set** which is the set of possible expected return and standard deviation pairs of all portfolios with different y (note that they all have the same Sharpe ratios)

-Naturally, an investor wants this line to be steeper since this would mean getting more return for more risk

-Can one go beyond the line where y = 1? Yes, by leveraging. But then the risk-free rate becomes modified because there is a risk of defaulting. Then the line becomes kinked beyond y = 1

## 6.5 Risk Tolerance and Asset Allocation

-Now that we have an equations of the expected returns portfolios with different capital allocation proportions, we can now link this with information on the risk aversion of the investor (the utility function)

we have

$$ U = E(r) - \frac{1}{2}A\sigma^2 $$

and

$$  E(r_c) = r_f + y(E(r_p) - r_f) $$

and

$$ \sigma_c^2 = y^2\sigma_p^2 $$

which gives:

$$ U = r_f + y(E(r_p) - r_f) - \frac{1}{2}Ay^{2}\sigma_p^2 $$

Solving the optimization problem 

$$ \frac{dU}{dy} = 0 $$

yields:

$$ y^{*} = \frac{E(r_p) - r_f}{A\sigma_p^2} $$

This shows that the optimal proportion of the risk asset is inversely proportional to the investor risk aversion and level of risk and directly proportional to the risk premium

## 6.6 Passive Strategies: The Capital Market Line

-A **Passive Strategy** avoids any direct or indirect security analysis

-Can choose the risky part of the portfolio to mimic the corporate sector of the US economy

  * e.g. the proportion of the capital invested in MSFT will be proportional to the ratio of the MSFT market capitalization to the total market capitalization. 

  * Such a portfolio earns a Sharpe ratio of .4 from the period of 1926-2015 (for every 1% of standard deviation, investors enjoyed .4% excess return or risk premium)

  * But from period to period, the Sharpe ratio varies a lot (by more than a factor of 2) therefore be careful when using historical data to forecast future returns

-The Capital Allocation Line from a portfolio of 1-month T-bills and a broad index of common stocks is known as the **Capital Market Line (CML)**

-Passive strategies may be better than an active one because of the lower costs and the free-rider benefit (active strategies will quickly eliminate inefficiencies in price)

## 7.1 Diversification and Portfolio Risk

-There are two broad sources of uncertainty

  1.) risk that comes from the general economy: the business cycle, inflation, interest rates, and exchange rates

  2.) risk that comes from the firm itself: success in R&D and personell changes (**nonsystematic risk**)

-Diversification reduces the risk of the portfolio but it cannot reduce the risk below the **systematic risk** or **nondiversifiable risk**

-Interestingly enough, even random diversification reduces risk!

-Risk reduction by spreading exposure across many independent risk sources is sometimes called the **insurance principle**

## 7.2 Portfolios of Two Risky Assets

-Now time to discuss *efficient* diversification where we construct risky portfolios which provide the lowest risk for a given expected return

-Consider a portfolio consisting of a bond portfolio (long-term debt securities), denoted D, and a stock fund specializing in equities, denoted E

-The return of such a portfolio:

$$ r_p = w_{D}r_{D} + w_{E}r_{E} $$

-The variance of a portfolio of two risky assets utilizes a formula from probability which is the variance of the sum of two random variables:

$$ \sigma_p^{2} = w_D^{2}\sigma_D^{2} + w_E^{2}\sigma_E^{2} + 2w_{D}w_{E}Cov(r_D,r_E) $$

-Note that this variance is less than the simple weighted sum of the variances, since the weights (<1) are squared which makes them smaller.
