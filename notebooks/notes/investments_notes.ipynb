## **1.1 Real Assets vs. Financial Assets**

-Wealth of a society is determined by the productive capacity of its economy aka the goods and services its member can create!

-The capacity is a function of the **real assets** of the economy: the land, buildings, machines, and knowledge that can be used to produce goods and services

-**Financial assets** do not contribute to the productive capacity of a society but they are claims on the real assets that do. They are claims on the income generated from these real assets!!!

-Buy shares and share in the income generated from using said real assets!

-Firms use this money to buy more real assets

-When aggregating all of the balance sheets in the economy, the financial assets and liabilities cancel each other and the net wealth of the economy is all real assets!

-Success or failure of financial assets ultimately depends on performance of underlying real assets!!!

## **1.2 Financial Assets**

-3 broad types of financial assets:

1.) fixed income

2.) equity

3.) derivatives

-**Fixed Income** (debt securities) promises fixed stream of income. e.g. corporate bond promises fixed interest. So least tied to performance of issuer (unless they go bankrupt!)

   * Money Market refers to debt securities that are short-term, highly marketable, and low-risk

     * e.g. Treasury bills or Bank Certificate of Deposit (CDs) 
  
   * fixed income capital market includes long-term securities

     * e.g. Treasury bonds as well as bonds issued by federal agencies, corporations, municipalities

-**Equity** represent ownership share in the corporation. Not promised any payment. Directly tied to success of the firm and its real assets and therefore its riskier

-**Derivative Securities** such as options and futures contracts provide payoffs determined by prices of other assets such as bonds or stocks. Named so because their value derives from value of other financial assets.
One use of derivatives is to hedge risks, aka transfer them to other parties

## **2.1 The Money Market**

-short-term debt securities, highly marketable, trade in large portions

-e.g. T-bills, money market mutual funds (accessible to small investors)

-T-bills

  * one way government raises money
  * buyer buys at discount and at maturity is paid the face value (par value)
  * maturities range from 4 weeks to 52 weeks
  * sell at denominations of $\$$100 but $10000 more common
  * Income earned from them exempt from state and local taxes!
  
  * Financial press reports **yields** for both **bid price** and **ask price**
  * ask price is what you would pay to buy a T-bill from securities dealer
  * bid price is the slightly lower price that you would sell to dealer
  * The **bid-ask spread** is the difference and it is the dealer's source of profit
  * The bid price is like the maximum willingness to pay for the dealer
  * The ask price is like the minimum willingness to sell for the dealer
  * When the bid-ask spread is narrow, it means the security is very liquid
  * Inverse relationship between the price and the yield (see https://xtbs.com.au/insights/price-yield-inverse-relationship/)
  * Also see Fidelity for a great explanation on relationship between interest rates, treasure prices, and yields (https://www.fidelity.com/learning-center/investment-products/fixed-income-bonds/bond-prices-rates-yields)

## **2.2 The Bond Market**

-longer term borrowing which includes Treasury notes and bonds, corporate bonds, municipal bonds, mortgage securities

### Treasure Notes and Bonds

-longer term (10 - 30 years)

-offer coupon payments (semiannual interest payments)

-prices quoted as percentage of par ($1000). See below figure.

-yield to maturity is the yield if all the coupons are reinvested

-**Yield** looks at change in interest or dividends and does not include capital gains (increase in security price) 

  * Obviously yield can only be applied to a narrow range of securities
  * return includes interest, dividends, and capital gains
  * There several different types of yield: **coupon rate**, **nominal yield**, **current yield**, and **yield-to-maturity**
  * **Current yield** is an investment's annual income (interest and dividends; in the case of bonds its interest) divided by its current price (which fluctuates)
  * **Nominal yield** is an investment's annual income (interest and dividends) divided by its par value
  * The **Coupon rate** is the **Nominal yield** and it is fixed!
  * Investors who pay a premium for a bond above its par value, will earn a rate of return thats less than nominal yield
  * Nominal yield (aka coupon rate) depends on inflation, real interest rate, and credit risk of issuer
  * Nominal yield is the rate of inflation plus the real interest rate (credit risky issuers also have higher yields)

## **2.3 Equity Securities**

### Common Stock as Ownership Shares

-*Common stock* aka *equity securities* aka *equities* are ownership shares in a corporation

  * Common because owners don't own any particular part of assets but rather the assets are common to all
  * each share entitles you to 1 vote on any matters of corporate governance that are put to the vote at annual meeting
  * also entitles you to share of any financial benefits
  * corporation is controlled by board of directors elected by shareholders
  * the board selects managers to run the corporation and makes sure they are in line with shareholder's interests
  * managers usually control annual vote by proxy vote (people who don't attend give their votes away)

### Characteristics of Common Stock

-**Residual Claim**
  * Shareholders have final claims on assets and income. If there is a liquidation, tax authorities, employees, suppliers, bondholders, and other creditors are paid first
  * In normal times, shareholders have final claims on operating income after interest and tax is paid. 
  * Management can pay out this residual through dividends or reinvest and make the share price go up
  
-**Limited Liability**
  * creditors can't come after shareholders' personal assets. For unincorporated businesses, creditors can come after owners' personal assets

### Stock Market Listings

-Listings include Close, Dividend (paid quarterly), Yield (annual dividend per dollar paid for the stock...remember yield does not include capital gains)

-**Capital Gains** are part of the return generated from price increases or decreases

  * low dividend firms offer greater potential for capital gains otherwise who would buy?
  
-**Price-Earnings Ratio** is the ratio of the current stock price to last year's earnings per share

  * how much you have to pay per dollar of earnings
  
-Where yield and P/E ratio not reported, company doesn't distribute dividends and has zero or negative earnings

### Preferred Stock

-Like debt security in the sense that its fixed income and you don't get voting rights

-Like an equity security in that firm retains discretion to pay the dividend

-Lower yields than bonds

### Stock Market Indexes

### **Dow Jones Industrial Average**

-30 large "blue-chip" corporations index computed since 1896

-Used to be a simple average of the prices

-Because of this summing of prices, it is **price-weighted**

-But the divisor is no longer 30

  * this divisor is adjusted for stock splits, dividend payments of more than 10%, or when one company in the 30 is replaced by another so as to keep the index continuous
  
  $$\text{DJIA} = {\sum p \over d}$$
  
  $$\text{DJIA} = {\sum p_\text{old} \over d_\text{old} } = {\sum p_\text{new} \over d_\text{new} }$$
  
  where p are the prices of the component stocks and d is the Dow Divisor
  
-Tries to be representative of the broad market

### **Standard & Poor's 500 Index**

-Better than DJIA because its 500 companies and its market cap weighted or **value-weighted**

-Essentially an average of the market capitalization of these 500 companies (market cap of publicly traded stocks! (free float))

-The divisor is once again adjusted to keep the S&P 500 continuous (but stock splits do not affect the market caps!)
  * additional share issuance, stock buyback, special dividends, etc. affects the divisor
  
  $$ \text{Index Level} = {\sum \left({P_i} \cdot {Q_i}\right) \over Divisor}$$
  
-Beauty of S&P 500 is that it reflects a straightforward portfolio strategy, namely, buy shares of the S&P 500 in proportion to their market capitalization. The index tracks the capital gains of this portfolio (see below calculation and plot)

-can invest through index funds or ETFs

## **3.2 How Securities Are Traded**

-Financial markets develop to meet the needs of particular traders

-imagine somebody wishing to invest in a financial asset would have to find someone willing to sell

   * eventually these people would find each other and meetups at a certain venue (this is how modern markets were formed)

### Types of Markets

4 types of markets: direct search, brokered, dealer, and auction

### Direct Search Markets

-least organized and buyers and sellers find each other through direct search (craigslist, facebook marketplace)

-participation is sporadic

### Brokered Markets

-next level of organization

-where participation is more active, people will pay for search services (broker)

   * this happens in real estate
   * brokers also develop knowledge in valuing assets traded in the market
   * primary market is also an example of a brokered market where IB is the one finding a buyer

### Dealer Markets

-when activity gets to the next level (number of transactions I suppose) for a particular type of asset, dealers emerge

-dealers buy the asset and then later sell them for profit 

   * the bid-ask spread is their profit
    
   * they save traders search costs because they can easily look up dealer's prices
    
   * there needs to be a lot of market activity before dealing is an attractive source of income

### Auction Markets

-most integrated market

-traders converge on one location either physically or electronically

   * NYSE is an example of an auction market
   * one doesn't need to search across dealers to find the best price
   * participants can save bid-ask spread (really? not my experience)

### Types of Orders

2 types of orders: market orders and orders contingent on price

### Market Orders

-market orders are buy or sell orders to be executed immediately at the current market prices

-you're quoted a bid and ask price

   * you'll pay the ask price (thats the minimum amount they're willing to sell you)
   * and if you sell it back, you'll get it at the bid price
   * These bid and ask prices are good only up to a certain amount of shares (**depth**)
   * After you exceed the depth, you might get charged more for the remaining than what you were quoted!
   * **depth** is a component of liquidity (its larger in S&P 500 than Russell 2000 (smaller stocks))
   * Another trader can beat you to the quote!
   * Even more, the quote can change before the order arrives, causing execution at a different price than the one quoted

### Price-Contingent Orders

-limit buy order specifies maximum to buy stock at

-limit sell order specifies minimum to sell stock at

-collection of **limit orders** is a *limit order book* (see below)

   * best orders at the top (highest limit buy order on top then descends, lowest limit sell order on top then ascends) 
   * best limit orders are the *inside quotes*
   * the *inside quotes* have small order sizes however
   * an investor looking to purchase larger order size will face an *effectice* bid-ask spread

## **3.5 New Trading Strategies**

-use of computer programs to make trading decisions

## **3.8 Buying on Margin**

-investors have access to a form of debt financing called *broker's call loans* and using this is called *buying on margin*

-buying on margin means the trader purchases a portion (the margin) of the price and borrows the rest

## **5.4 Risk and Risk Premiums**

### Holding Period Return

-Essentially the return you get during a certain investment horizon

$$ \text{HPR} = \frac{\text{Ending price of a share - Beginning price + Cash dividend}}{\text{Beginning price}} $$

-Essentially, the return on equity is the sum of the **dividend yield** and **capital gains** 

-percentage return from dividend is the **dividend yield**

### Expected Return and Standard Deviation

-if you have an investment with a certain horizon and probability distribution of HPR at that time, you can calculate the **expected return**
  * Just the expectation value of the return over this pdf of returns at the end of the time horizon
  * Can also calculate the volatility and standard deviation with this probability distribution and expectation value

### Excess Returns and Risk Premiums

-How much would you invest in a financial asset?

  * Depends on the expected reward for the given risk
  
-Measure the reward as the difference in the *expected* HPR and the **risk-free rate** (rate earned from risk-free assets such as T-bills, money market funds, or the bank

  * We call this the **risk premium**
  
  $$ \text{Risk Premium} = E[r] - r_{f} = E[r - r_{f}] $$
  
  
  * The difference betrween the *actual* return and the risk-free rate is the **excess return**
  
  $$ \text{Excess Return} = r - r_{f} $$
  
  * The risk is the standard deviation of the excess return
  
  $$ \text{Risk} = \sigma[r - r_{f}] = \sqrt{var[(r - r_f)]} $$
  
-**risk averse** investors require a risk premium to invest

-Note: The benchmark risk-free rate (should be real) should have a maturity which matches the investment horizon! 


## **5.5 Time Series Analysis of Past Rates of Return**

### Time Series vs. Scenario Analysis

-In forward-looking scenario analysis (above), we determine a set of scenarios with associated HPRs and probabilities and calculate the risk premium (reward) and standard deviation (risk)

-In contrast, asset return histories do not come with the probabilities at the time of observation and we must infer the probability distributions 

### Expected Returns and the Arithmetic Average

-Equally weight each observation and simply do an average to calculate the expected return

  * arithmetic average equivalent to summing over a uniform probability distribution

### The Geometric (Time-weighted) Average Return

-arithmetic tells us about the *expected* return but what about the *actual* performance? 

-This is essentially the EAR (you're uncompounding to get the annual)

-In general, the geometric average of returns is less than the arithmetic average of returns, and if the returns are normal, the difference is:

$$ E[\text{Geometric average}] = E[\text{Arithmetic average}] - \frac{1}{2}\sigma^{2} $$

### Variance and Standard Deviation

-With regards to risk, we are interested in the deviations of actual outcomes from the *expected* return

  * but we cannot measure the expected return so we estimate it with the arithmetic average
  
  $$ \hat{\sigma}^2 = \frac{1}{n} \sum_{s=1}^{n} [r(s)-\bar{r}] $$
  
-However, this estimate is biased downward, because we have taken deviations from the sample arithmetic average instead of the unknown expectation value of the returns

  * To fix, simply multiply by (becomes relevant for large sample sizes) $$ \frac{n}{n-1} $$

### ***Mean and Standard Deviation Estimates from Higher-Frequency Observations**

-frequency of observations does not improve estimate of expected return!!!

  * intra-year observations, for example, provide no additional information about mean return, since they have to be consistent with the yearly return

-duration of observations does improve estimate accuracy of estimate of expected return

  * This suggests a rule: Use the longest sample that you believe comes from the same return distribution!!

-But the accuracy of estimates of the standard deviation can be increased with increased number of observations

  * more frequent observations gives us more information about the distribution of deviations from the average
  
-Estimates of the standard deviation begin with the variance. *When the returns are uncorrelated*, the variances add up

$$ \sigma_A = \sqrt{12}\sigma_M $$

### The Reward-to-Volatility (Sharpe) Ratio

-Essentially take the ratio of the reward (risk premium) to the risk (standard deviation of the excess return)

$$ \text{Sharpe Ratio} = \frac{E[R_a - R_f]}{\sqrt{var[R_a - R_f]}} $$

-Used to evaluate the performance of investment managers

-Adequate measure of the risk-return trade-off for diversified portfolios but inadequate for individual assets such as shares of stock!!!

## **5.7 Deviations From Normality and Alternative Risk Measures**

-deviations from normality measured by greater than second order moments of the distribution

-skewness, the third moment, is the expectation of the cubed deviations from the mean over the cubed standard deviation

$$ \text{Skew} = E[\frac{(R-\bar{R})^3}{\hat{\sigma}^3}] $$

-When the returns are positively skewed, rare extreme positive events may occur and the distribution is "skewed" to the right

  * The standard deviation tends to overestimate the risk in this case! (Just like how the mean will overestimate the 50th percentile with positive skewness)
  * Likewise the standard deviation underestimates the risk with negatively skewed distributions
  
-kurtosis, the fourth moment, the expectation of the deviations from the mean over the standard deviation to the fourth power

$$ \text{Kurt} = E[\frac{(R-\bar{R})^4}{\hat{\sigma}^4}] - 3 $$

-This concerns the increased likelihood of extreme events on either side in the form of "fat tails"

-With high kurtosis, the standard deviation would underestimate the likelihood of extreme events

-Apparently skewness and kurtosis do not changes when annualized from high frequency observations!!!

### Value at Risk

-simply the loss corresponding to a very low percentile of the returns distribution

-Typically its the 1% quantile of the distribution (99% of the returns will exceed it)

### Expected Shortfall

-VaR is optimistic because it takes the highest values of that those returns lying below the 1% quantile

-The **expected shortfall** or **conditional tail expectation** is simply the expectation value of that worst 1% of returns and is more informative

### Lower Partial Standard Deviation and the Sortino Ratio

-issue with using standard deviation for risk is asymmetry in distributions and the fact that we should really be looking at deviations from the risk free rate and not the mean rate of return

-**Lower partial standard deviation (LPSD)** is essentially the standard deviation of the deviations from the risk-free rate, conditioned on it being a negative deviation

-When you replace the denominator of the Sharpe Ratio with LPSD, it is the **Sortino Ratio**

### Relative Frequency of Large, Negative 3-Sigma Returns

-Essentially compare the frequency of negative 3 sigma events to frequency of such events for a normal

## **5.9 Normality and Long-Term Investments**

****Even if short-term stock returns were normal, long term would not be!***
  * This is due to the fact that the returns do not add through time, they multiply (aka compound) 
  
  $$ (1 + r_1)(1 + r_2) \text{  not  } r_1 + r_2 $$
  
-In the long-run, the returns distribution begins to look lognormal with a positive skewness
  * the upside is unlimited but downside has a floor (you can't lose more than you invested
  
  * but using continuous compounding, the compounded returns are multiplied exponentials and the exponent then is additive in the returns so it stays normal (just have to log transform the returns)
  
  $$ \exp{r_{cc}(1)}*\exp{r_{cc}(2)} = \exp{r_{cc}(1) + r_{cc}(2)} $$
  
  * In addition, if the returns are the same over each holding period, then you just do n number of holding periods times the return
  * Also, if the returns are uncorrelated, the variance is also just n times the variance of the single holding period. 
  * generalization to arbitrary investment horizons is:
  
  $$ E(W_{T}) = W_{0}*\exp{E(r_{cc})T} $$
  
  Where $$ E(W_{T}) $$ is the expected final value of the portfolio
  
  Also, the variance is proportional to the time horizon:
  
  $$ Var(r_{cc}T) = TVar(r_{cc}) $$
  
  Therefore,
  
  $$ \sigma(r_{cc}T) = \sqrt{T}\sigma $$

### Short-Run versus Long-Run Risk

-This analysis appears to offer a relief from investment risk: **Because expected return increases faster with horizon than risk (std increases as sqrt of T while the expected return increases with T), you can argue that investments are less risky when horizons are large**

  * but worse case scenarios get worse, even though less likely
  * The 1% VaR grows with investment horizon so the *probability* of losses falls with investment horizon but the *magnitude* of the potential losses grows
  
-better way of quantifying risk of long-term investment is to calculate the market price of insuring it against a loss

-speculation is in spite of the considerable risk because you perceive a favorable risk-return tradeoff
  * *risk aversion and speculation are consistent* (weird)
  * **fair game** means risk premium of zero and it amounts to a gamble (expected return is zero)
  * essentially gambles have a risk premium of 0 or below

### Risk Aversion and Utility Values

-risk tends to increase with increased risk premium... so how do you score? of course penalize higher risk and reward higher return

-Utility refers to how much benefit investors derive from portfolio performance

-common utility function is as follows:

$$ \boxed{U = E(r) - \frac{1}{2}A\sigma^2} $$

Where A is the investor's degree of risk aversion (how much their utility decreases with increasing risk)

-Investors select portfolio with highest utility! 

-**Certainty Equivalent Rate** is the rate that a risk-free investment would need to have to have the same utility as the portfolio
  * The investor would only choose the portfolio if its certainty equivalent rate is greater than the risk free rate
  * Essentially, it is the guaranteed return that the investor would accept now (since its utility is equal to or greater than the proposed portfolio)

-**Risk-neutral** investors (A=0) only judge a portfolio off of its expected returns

-**Risk lover** investors (A<0) happily engage in fair games and gambles

-One can plot expected return vs the risk. The northwest the direction is the preferred one since this increases expected return and reduces risk (for a risk averse investor)

-**mean-variance (M-V) criterion**
  * Portfolio A dominates B if:
  $$ E(r_A) >= E(r_B) $$
  and
  $$ \sigma_A <= \sigma_B $$

-Can find the **indifference curve** by setting the utility to a constant and plotting all of the expected return and standard deviation values which satisfy this

-You can take surveys which quantify your risk aversion!

## **6.4 Portfolios of One Risky Asset and a Risk-Free Asset**

-Keep in mind the difference between asset allocation and capital allocation
  * *Asset Allocation* is the adjustment of the weights of the risky part of the portfolio (the allotment of funds between stocks and funds)
  * *Capital Allocation* is the adjustment of the weights between the risk-free investments and the risky assets

-Let $y$ by the proportion of funds allocated to the risky asset and therefore $(1-y)$ be allocated to the risk-free asset. If the return on the risky asset is $r_p$ and the return on the risk-free asset is $r_f$ then the return on the complete portfolio is:

$$ r_c = yr_p + (1-y)r_f $$

And the expected return is:

$$ E(r_c) = yE(r_p) + (1-y)r_f $$

Rearranging, we arrive at an intuitive formula:

$$ E(r_c) = r_f + y(E(r_p) - r_f) $$

Essentially the base rate of return is the risk-free rate and the additional return is the proportion of the risky asset times the risk premium

Also, the standard deviation of the portfolio is essentially the std of the risky asset times its proportion (from stat, if you muiltiply a r.v. by a constant, the std is multiplied by that constant)

$$ \sigma_c = y\sigma_p $$

Using this, one can solve for the expected return of the portfolio as a function of the standard deviation

$$ \boxed{E(r_c) = r_f + \sigma_c(\frac{E(r_p) - r_f}{\sigma_p})} $$

Notice this is a line in the expected rate of return - standard deviation plane and the slope is the Sharpe Ratio. This line is what is referred to as the Capital Allocation Line (CAL)

-Essentially, this line is parametrized by y, the proportion allocated to the risky asset. Increasing y represents moving northeast on this line. 

```python
Image(filename='../../../data/CAL.png') 
```

-The above represents the **investment opportunity set** which is the set of possible expected return and standard deviation pairs of all portfolios with different y (note that they all have the same Sharpe ratios)

-Naturally, an investor wants this line to be steeper since this would mean getting more return for more risk

-Can one go beyond the line where y = 1? Yes, by leveraging. But then the risk-free rate becomes modified because there is a risk of defaulting. Then the line becomes kinked beyond y = 1

## **6.5 Risk Tolerance and Capital Allocation**

-**The capital allocation exercise is essentially one of optimizing the investor's utility, given the expected return of the portfolio as a function of its risk (parametrized by the proportion given to the risky assets. This function is the Capital Allocation Line) and given the investor's risk aversion.**

-Now that we have an equation of the expected returns portfolios with different capital allocation proportions, we can now link this with information on the risk aversion of the investor (the utility function)

we have

$$ U = E(r) - \frac{1}{2}A\sigma^2 $$

and

$$  E(r_c) = r_f + y(E(r_p) - r_f) $$

and

$$ \sigma_c^2 = y^2\sigma_p^2 $$

which gives:

$$ U = r_f + y(E(r_p) - r_f) - \frac{1}{2}Ay^{2}\sigma_p^2 $$

Solving the optimization problem 

$$ \frac{dU}{dy} = 0 $$

yields:

$$ \boxed{y^{*} = \frac{E(r_p) - r_f}{A\sigma_p^2}} $$

This shows that the optimal proportion of the risky asset is inversely proportional to the investor risk aversion and level of risk and directly proportional to the risk premium

-Now time to discuss *efficient* diversification where we construct risky portfolios which provide the lowest risk for a given expected return

-Consider a portfolio consisting of a bond portfolio (long-term debt securities), denoted D, and a stock fund specializing in equities, denoted E

-The return of such a portfolio:

$$ \boxed{r_p = w_{D}r_{D} + w_{E}r_{E}} $$

$$ \implies \boxed{E(r_{p}) = w_{D}E(r_{D}) + w_{E}E(r_{E})} $$

-The variance of a portfolio of two risky assets utilizes a formula from probability which is the variance of the sum of two random variables:

$$ \boxed{\sigma_p^{2} = w_D^{2}\sigma_D^{2} + w_E^{2}\sigma_E^{2} + 2w_{D}w_{E}Cov(r_D,r_E)} $$

-Note that this is always smaller than or equal to the weighted sum of the variances. To see this, see:

$$ Cov(r_{D},r_{E}) = \rho_{DE}\sigma_{D}\sigma_{E} <= \sigma_{D}\sigma_{E} $$

$$ \implies \sigma_{P}^{2} = w_{D}^{2}\sigma_{D}^2 + w_{E}^{2}\sigma_{E}^2 + 2w_{D}w_{E}\sigma_{D}\sigma_{E}\rho_{DE} $$

Since the correlation coefficient ranges from -1 to 1:

$$ \implies \sigma_{P}^{2} \leq w_{D}^{2}\sigma_{D}^2 + w_{E}^{2}\sigma_{E}^2 + 2w_{D}w_{E}\sigma_{D}\sigma_{E} $$

$$ \implies \sigma_{P}^{2} \leq (w_{D}\sigma_{D} + w_{E}\sigma_{E})^2 $$

$$ \implies \boxed{\sigma_{P} \leq (w_{D}\sigma_{D} + w_{E}\sigma_{E})} $$

-**A hedge asset has negative correlation with the other assets in the portfolio and from above equation, it can be seen how a negatively correlated asset would redue the variance of the portfolio.**

-**Futhermore, the equation for the expected return of the portfolio shows us that the correlation between the assets does not affect it!!!**

-**Therefore, portfolios of less than perfectly correlated assets offer some diversification benefit**

-How low can portfolio standard deviation be? extreme case is correlation coefficient is -1:

$$ \implies \sigma_{P}^{2} = w_{D}^{2}\sigma_{D}^2 + w_{E}^{2}\sigma_{E}^2 - 2w_{D}w_{E}\sigma_{D}\sigma_{E} $$

$$ \implies \sigma_{P}^{2} = (w_{D}\sigma_{D} - w_{E}\sigma_{E})^2 $$

$$ \implies \sigma_{P} = |w_{D}\sigma_{D} - w_{E}\sigma_{E}| $$

A **perfectly hedged position** can be obtained by choosing weights s.t.:

$$ \sigma_{P} = 0 $$

$$ \implies w_{D}\sigma_{D} - w_{E}\sigma_{E} = 0 $$

$$ \implies w_{E} = \frac{\sigma_{D}}{\sigma_{D} + \sigma_{E}} $$

$$ \text{and } w_{D} = \frac{\sigma_{E}}{\sigma_{D} + \sigma_{E}} $$



-Determining the **Minimum Variance portfolio** involves solving the optimization problem of minimizing the portfolio variance

-**Essentially, the task is to minimize the variance of the risky portfolio with respect to the weights of the risky assets**

$$ \sigma_p^{2} = w_D^{2}\sigma_D^{2} + w_E^{2}\sigma_E^{2} + 2w_{D}w_{E}Cov(r_D,r_E) $$

$$ \frac{\partial{\sigma_{p}}}{\partial{w_{p}}} = 0 $$

$$ \implies w_{E}^{*} = \frac{\sigma_{D}^2 - \sigma_{D}\sigma_{E}\rho_{DE}}{\sigma_{D}^2 + \sigma_{E}^2 - 2\sigma_{D}\sigma_{E}\rho_{DE}}$$

-Recall the forms of the portfolio variance for perfectly anti-correlated, partially correlated, and perfectly correlated:

**Perfectly Anti-Correlated**:

$$ \sigma_{P} = |w_{D}\sigma_{D} - w_{E}\sigma_{E}| $$

**Partially Correlated**:

$$ \sigma_{P}^{2} = w_{D}^{2}\sigma_{D}^2 + w_{E}^{2}\sigma_{E}^2 + 2w_{D}w_{E}\sigma_{D}\sigma_{E}\rho_{DE} $$

**Perfectly Correlated**:

$$ \sigma_{P} = (w_{D}\sigma_{D} + w_{E}\sigma_{E}) $$

These are simply lines and a polynomial in the standard deviation - stock weight plane:

  * However, if the correlation is perfect, it is a straight line and no diversification benefit is possible

## **7.4 The Markowitz Portfolio Optimization Model**

-The first step is the construction of the **minimum variance frontier**, which is the lowest possible variance that can be attained for a given *portfolio* expected return

  * if you look at equations for the portfolio expected return of multiple risky assets, the portfolio variance of multiple risky assets and the optimization equations for the variance, in theory you should be able to solve for each point on the minimum variance frontier **given a portfolio expected return target** 

-All of the individual securities lie to the right of the minimum variance frontier (if short selling is allowed) thus illustrating effect of diversification

-All of the portfolios upward from the global minimum variance portfolio are the preferred portfolios (Recall Mean-Variance Criterion) and this is what we call the **Efficient Frontier**

  * The bottom part of the curve has a portfolio with the same standard deviation but higher expected return on the efficient frontier, hence the bottom part is inefficient

```python
Image(filename='../../../data/efficientfrontier.png') 
```

-The optimal risk portfolio is then found by finding the tangent portfolio, aka the one which maximizes the slope of the CAL. Or more intuitively, the one which maximizes the Sharpe Ratio.

-Finally, you optimize the investor's utility and find the proper mix between the risky portfolio and T-bills

-Generalization of our equations for n risky assets is then:

$$ E(r_{p}) = \sum_{i=1}^{n} w_{i}E(r_{i}) $$

$$ \sigma_{p}^2 = \sum_{i=1}^{n}\sum_{j=1}^{n} w_{i}w_{j}Cov(r_{i},r_{j}) $$
  

-Markowitz won the Nobel Price in Economics in this, and it is essentially step one of portfolio management: the identification of the efficient set of portfolios, or the *efficient frontier of risky assets*

-Practically, your input list is an estimate of the expected rates of return and covariance matrix of the securities. The optimization algorithm then finds the efficient frontier. 

  - There may be additional constraints to the weights adding to 1. Some clients do not what short sells, some clients want a minimum divident yield and others maybe want to stay out of securities from unstable countries. Additional constraints lower potential Sharpe ratios. 

### Capital Allocation and the Separation Property

-You can either find the tangency portfolio by plotting the CAL and ratcheting it up, or you can simply maximize the Sharpe ratio

-**The interesting implication here is that the portfolio manager will offer the same risky portfolio to all clients, regardless of their risk aversion! (additional constraints being equal)** 
  - this makes investment management less costly as they can serve a large variety of clients

-The **Separation Property** is essentially the fact that constructing the investor tailored optimal portfolio can be separated into two tasks: one of the technical optimization of the risky portfolio and one of the investor-specific capital allocation step. 

-In practice, different managers will come up with different input lists ($E(r) \text{ and cov}$) and hence offer different optimal portolios
  - the difference is in the security analysis! If the expected returns are miscalculated resulting from mispricing, you'll have garbage in, garbage out
  - a crappy input list would be one which equated expected returns with the historic returns. 
  
    - I.E. if you used a 10 year average to estimate the expected return of a stock with an annual standard deviation of 50%
      - From statistics, we know that the standard deviation of this sample mean is: 
      $$ \sigma_{mean} = \frac{1}{\sqrt{N}}\sigma = 50 / \sqrt{10} = 15.8 \% $$
    
      - This is on the order of magnitude of the return itself!!!
      - **and this is not improved by sampling more frequently!! (See the end of chapter 5)** 

-Nevertheless, a limited number of portfolios will serve a wide range of clients, and this is the theoretical basic of the mutual fund industry!

### The Power of Diversification

-The limit of diversification in the limit of $n$ risky assets is the systematic risk of the general economy

-Recall the variance of a portfolio of $n$ risk assets:

$$ \sigma_{p}^2 = \sum_{i=1}^{n}\sum_{j=1}^{n} w_{i}w_{j}Cov(r_{i},r_{j}) $$

-Now consider a naive diversification strategy of equal weighting where each weight is $1/n$

-This equation becomes:

$$ \sigma_{p}^2 = \frac{1}{n}\sum_{i=1}^{n}\frac{1}{n}\sigma_{i}^2 + \sum_{\substack{j=1\\j\not=i}}^n\sum_{i=1}^n\frac{1}{n^2}Cov(r_{i},r_{j}) $$

-You can then define an average variance and covariance of the securities:

$$ \overline{\sigma}^2 = \frac{1}{n}\sum_{i=1}^{n}\sigma_{i}^2 $$
$$ \overline{Cov} = \frac{1}{n(n-1)}\sum_{\substack{j=1\\j\not=i}}^n\sum_{i=1}^n\frac{1}{n^2}Cov(r_{i},r_{j}) $$

-So that the portolio variance becomes: 

$$ \sigma_{p}^2 = \frac{1}{n}\overline{\sigma}^2 + \frac{n-1}{n}\overline{Cov} $$

-Now examine the effect of diversification:

  - as n becomes larger, the first term disappears and you're left with the second term which would disappear if the securities are uncorrelated
  - However, when economy-wide risk factors exist, securities are necessarily correlated
    - Thus the irreducible risk comes down to the covariance of the securities which depends on the systematic variance of the macroeconomic variables.
  - But the greater diversification effect of reduction in portfolio variance comes with smaller correlation between the securities
  - **The most important insight here is that with diversified portfolios, what ultimately drives reduction in risk is the covariance of the securities and not necessarily the individual variances**

## **7.5 Risk Pooling, Risk Sharing, and the Risk of Long-Term Investments** 

-Risk reduction actually involves both **risk pooling** (spreading exposure across multiple uncorrelated risky assets) and **risk sharing** (allowing other investors to share in the risk of a portfolio of assets)
  - The utility of capital markets in a developed economy rests on this distinction

### Risk Pooling and the Insurance Principle

-essentially adding risky projects to the investor's portfolio

-But even brief reflection on this should convince you that this is not the whole story. How can adding more bets that are independent of your other bets reduce your total exposure to risk?

  - Like if a gambler in Las Vegas took a few more trips to the roulette table to reduce his total risk by diversifying his overall "portfolio" of wagers

    - The reason this is not so is that while his average gain per bet may become more predictable as he repeatedly returns, his total proceeds become less so

    - if you apply the same probability theory formulas to the variance of the *dollar* profit, you will find that the variance increases (since we don't have the weight factors which come from the rate of return analysis. see pg.218 of Zvi to get a great example.)

    - so for the gambler, while his *percentage* gains may be more predictable with every trip, his *dollar* gains are less so

### Risk Sharing

- Now imagine that investor/gambler sold off part of his wagers to other investors and kept his total investment at the initial wager, even though he added a second wager to the portfolio
  - so now we can apply our equations of expected rate of return and the variance to the gains without worrying about the increasing variance to the gains as we scale up

-The insurance industry runs off of the insurance principle (risk pooling), however a key part is risk sharing
  - risky pooling makes the variance of the rate of return less uncertain, but the uncertainty in the profits larger
  - but by bringing in shareholders, who own a piece of those investments, the risk to the profits actually goes down

### Diversification and the Sharpe Ratio

-While risk pooling by itself doesn't by itself reduce risk, it does not mean it is not unimportant

-risk pooling improves the Sharpe ratio, regardless of whether risk sharing is part of it

-so going back to the gambler, if he took 100 bets, they are certainly not less risky than one bet. But if he shared those 100 bets with 100 friends, it is indeed less risky

-**True diversification is spreading a portfolio of *fixed size* across many assets, not merely adding more risky bets to an ever-growing risky portfolio**

### Time Diversification and the Investment Horizon

-now think of extending the time horizon as adding more risky assets
  - while it does improve the Sharpe ratio, it also increases the risk

## **8.1 A Single-Factor Security Market**

### The Input List of the Markowitz Model

-The number of estimates of returns scales as $n$, the number of instruments, the number of estimates of variances scales as $n$, while the number of estimates of covariances increase quadratically with the number of instruments:

$$ (n^2 - n)/2 $$

 - 50 instruments implies 1,325 total estimates, while 3,000 instruments implies 4.5 million estimates!!!
 - crappy estimates can screw up your optimization

-Another issue is that an inconsistent correlation matrix can lead to nonsensical variance calculations

-The simplification with the index models comes with the realization that correlations among securities arise from economic forces that affect the productivity of firms (e.g. business cycles, interest rates, and the cost of natural resources. Shocks cause correlated shocks in the rates of return on stocks across the entire market)

  - essentially decompose the uncertainty into systemwide and firm-specific sources

### Systematic versus Firm-Specific Risk

-Separate risk into sum of previously expected value plus an unanticipated surprise:

$$ r_i = E(r_i) + \text{unanticipated surprise} $$

-unanticipated surprise could be due to firm or broader economy

 - broaded economic uncertainty: $ m $, and firm-specific uncertainty denoted by: $ e_i $

-**The common dependence that all securities have to the broader economy is the source of the correlation between their security returns!!**

-Both $m$ and $e_i$ are assumed to have zero mean (surprises average out to zero) and they are assumed to be uncorrelated
  - Thus the variance in $r_i$ arises from two uncorrelated sources: the firm and the economy

-We can capture different firms sensitivity to macroeconomic conditions with $\beta_i$ and then write the **single factor model**:

$$ r_i = E(r_i) + \beta_i m + e_i $$

-Since $m$ and $e_i$ are uncorrelated, the variance is:

$$ \sigma_i^2 = \beta_i^2\sigma_m^2 + \sigma_{e_i}^2 $$

-Where $\beta_i^2\sigma_m^2$ is what is known as the systematic risk and is obviously higher when $\beta_i$ is higher!

-For a portfolio of securities, we note the assumption that the firm specific surprises are mutually uncorrelated, so that the covariance is simply due to the firms' correlations to the broader market:

$$ Cov(r_i,r_j) = Cov(\beta_i m + e_i, \beta_j m + e_j) = \beta_i\beta_j\sigma_m^2 $$

-We need a variable which can serve as a proxy for this dependence on the broader economy!

## **8.2 The Single-Index Model**

-Since the systematic factor affects the rate of return on all stocks, the rate of return of a broader market index can proxy for that common factor

-For the single-index model, we cast the single-factor model as a regression of the security excess return onto the market index excess return:

$$R_i(t) = \alpha_i + \beta_i R_M(t) + e_i(t) $$

Where $R_i(t) = r_i(t) - r_f$ and $R_M = r_M - r_f$ are the excess returns of the security and index, respectively. $\alpha_i$ is the security's excess return when the market index excess return is zero. $\beta_i$ is of course the sensitivity of the security to the index and is the slope of the regression line, while $e_i(t)$ is the zero-mean firm-specific surprise known as the residual

-The average beta of all stocks in the economy is 1. 
  - Aggressive stocks have beta greater than 1 (per unit change of the index, the aggressive stocks change that much more)
  - "Defensive" have beta less than 1 (per unit change of the index, the defensive stocks change that much less) 

### The Expected Return-Beta Relationship

-Since $E_(e_i) = 0$, we have:

$$ E(R_i) = \alpha_i + \beta_i E(R_M) $$

-So part of the security's risk premium is due to the risk premium of the index! 
  - Recall  $ \text{Risk Premium} = E[r] - r_{f} = E[r - r_{f}] $
  - The market risk premium is multiplied by beta, which makes sense because securities which are more sensitive to market risk compensate that with a higher risk premium (called the *systematic risk premium*)
  - \alpha is the nonmarket premium, which might be due to the underpricing of the security

### Risk and Covariance in the Single-Index Model

-The number of estimates now scales linearly with the number of securities in the portfolio due to the decomposition of firm-specific and market-specific risk

-This decomposition also helps with specialization of sectors and industry-specific knowledge in the securities analysis
  - essentially, managers at an investment company don't need to worry about covariance between computer industry and auto industry
  - they simply need to understand covarance between their securities and broader market index

-This oversimplification (essentially the classifiction of uncertainty into the simply dichotomy of micro vs macro risk) of the model rules out industry wide events, which affect the firms within that industry but not the economy at large


### The Index Model and Diversification

-Assume an equally weighted portfolio. Then our Index model becomes:

$$ R_P = \alpha_P + \beta_P R_M + e_P $$

Where $\alpha_P = \frac{1}{n}\sum_{i=1}^{n}\alpha_i$, $\beta_P = \frac{1}{n}\sum_{i=1}^{n}\beta_i$, $e_P = \frac{1}{n}\sum_{i=1}^{n}e_i $ and the covariance is:

$$ \sigma_P^2 = \beta_P^2\sigma_M^2 + \sigma^2(e_P) $$

Where $\alpha$ does not have a variance because I believe it is assumed to not have a random variable. One can see from this equation that as $n$, the number of securities, in the portfolio grows, the right term will go to zero due to law of large numbers, but the left term will persist.

-Esssentially, when diversification increases, the total variance of the portfolio approaches the systematic variance
  - its why we say systematic risk is nondiversifiable

## **8.3 Estimating the Single-Index Model**

### The Security Characteristic Line for Ford

-The SCL of Ford is simply the line determined by the index model applied to Ford
-Follow the standard statistical regression techniques to estimate parameters (Multiple R (essentially correlation), $R^2$, Adjusted $R^2$, coefficient standard errors and p-values)):

$$ R_{Ford}(t) = \alpha_{Ford} + \beta_{Ford}R_{index}(t) + e_{Ford}(t) $$

  - firm-specific risk typically tends to be larger than the systematic risk
  - firms with high exposure to the state of the macroeconomy, will have $\beta$ greater than 1 (US Steel, Ford, Disney)

## **8.5 Portfolio Construction Using the Single-Index Model (Treynor-Black Model)**

-The goal is the formation of portfolios with efficient risk-return trade-offs, but simplifications offered by index model change the procedure

### Alpha and Security Analysis

-Using the Markowitz model, we need estimates of the risk premiums of each security, which depends on the macroeconomic and firm-specific forecasts
  - In a large organization, this would pose a problem when analysts for different sectors possibly have different macro forecasts
  - Index models separates these two sources of return variation

-Procedure:
  - Use macroeconomic analysis to estimate the risk premium and risk of the market index
  - Use statistical analysis to estimate the $\beta$ coefficients of the securities and their residual variances $\sigma^2(e_i)$
  - The manager then uses these estimates to come up with the purely market-driven expected return
  - Security-specific expected return forecasts (the security alphas) are derived from various security-valuation models. This gives us the *incremental* risk premium attributable to private information developed from security analysis

-Statistical methods of estimating $\beta$ coefficient are widely known and standardized (won't differ across portfolio managers)

-In contrast, macro analysis and security analysis are far less routine and provide the potential for distinguished edge

-$\alpha$ is the key variable which tells us whether a security is a good or bad buy
  - For any given $\beta$ which we determine, we can easily find another security with a similar correlation to the market index. Therefore, what really makes a security attractive is its premium above and beyond its market premium
  - a positive $\alpha$ instrument is a bargain and should be overweighted in the overall portfolio 

### The Index Portfolio as an Investment Asset

-The investment manager might include the S&P 500 portfolio as one of the assets in the portfolio

  - This is due to the fact that the prospectus of an investment company might limit the universe of assets in which the firm may invest to only stocks included in the S&P 500
  - Furthermore, resource contraints might limit this investable universe to a subset of the S&P 500
  - Because of this, there might be a worry of limited diversification, so you throw in an S&P 500 portfolio

### The Single-Index Model Input List

-Risk premium of the S&P 500 portfolio

-Standard deviation of the S&P 500 portfolio

-*n* sets of estimates of beta coefficients, stock residual variances, and alpha values

### The Optimal Risky Portfolio in the Single-Index Model (Treynor-Black)

-Analogous to the previous example of the equally weighted portfolio, it turns out that alpha, beta and the variance of the firm-specific residuals gets a simple modification:

$$\alpha_P = \sum_{i=1}^{n+1}w_{i}\alpha_i$$ 

$$\beta_P = \sum_{i=1}^{n+1}w_{i}\beta_i$$

$$\sigma^2(e_P) = \sum_{i=1}^{n+1}w_{i}^2\sigma^2(e_{i}) $$

where $w$ are the weights to be determined by the optimization and $n+1$ is the S&P 500 portfolio

-**The optimal portfolio turns out to be a combination of an *active portfolio* (comprised of n securities analyzed through securities analysis) and a *passive portfolio* (a portfolio of the S&P 500 for example)**

### The Information Ratio

-The Sharpe ratio of the optimizally constructed risky portfolio will exceed that of the index portfolio:

$$ S_P^2 = S_M^2 + [\frac{\alpha_A}{\sigma(e_A)}]^2 $$

This is known as the **information ratio** and it measures the extra return we can obtain from security analysis compared to the firm-specific risk we incur when we over or underweight securities relative to the passive market index

-it turns out that the information ratio will be maximized if we invest in each security in proportion to its ratio $\frac{\alpha_i}{\sigma(e_i)}$

  -**It was shown by Treynor and Black that the active portfolio is constructed using the weights:**

$$ w_i = \frac{\alpha_{i}/\sigma_{i}^2}{\sum_{j=1}^{N}\alpha_{j}/\sigma_{j}^2} $$

$$ [\frac{\alpha_A}{\sigma(e_A)}]^2 = \sum_{1=1}^{n}[\frac{\alpha_i}{\sigma(e_i)}]^2 $$

-A further note: the index portfolio is the efficient portfolio only if all alpha values are zero

  - This is why its all about finding alpha!!!

# **The Capital Asset Pricing Model**

-one of the centerpieces of modern finance

-gives us a prediction of the relationship between the risk of an asset and its expected return

  - This serves the purpose of setting a benchmark rate of return for possible investments
  - This also serves as an educated guess of assets which have not been traded publicly yet

-However, CAPM does not withstand empirical tests!

-But its central conclusion is that only systematic risk will be rewarded with a risk premium

## **9.1 The Capital Asset Pricing Model**

-assumptions

  - all investors are mean-variance optimizers with a common time horizon and common set of information
  - markets are well-functioning with few impediments

-The central implication of the model is that risk premia are proportional to systematic risk and independent of firm risk

-Because of all the stringent assumptions, all of the investors arrive at the same optimal portfolio

  - but because the market portfolio (market cap weighted portfolio of all the assets in the universe) is an aggregation of these identical investor portfolios, the investor portfolios must be the market portfolio!

  - As a result, each investor portfolio is simply a share of the market portfolio

### The Passive Strategy is Efficient

-According to CAPM, the CML (capital market line) is the optimal tangency portfolio on the efficient frontier

-So investors can skip security analysis and simply hold the market portfolio!

  - Of course, this is paradoxical because if everyone did this, nobody would perform security analysis

-But still, the academics say that the passive strategy of investing in a market-index portfolio is efficient

  - This is the **mutual fund theorem** (incarnation of separation property of chapter 7)

### The Risk Premium of the Market Portfolio

-recalling the solution for the proportion allocated to the optimal portfolio, we have:

$$ y = \frac{E(r_M) - r_f}{A\sigma_{M}^2} $$

-The average position in the risky portfolio is 100% (not sure why) so that the risk premia on the market portfolio is:

$$ E(R_M) = \bar{A}\sigma_M^2 $$

Where $\bar{A}$ is the representative investor's risk aversion

### Expected Returns on Individual Securities

-Portfolio risk is what governs the risk premiums which investors demand

-It can be shown (pg. 283 of Investments) that an individual stock's contribution to the variance of the portfolio is:

$$ w_{GE}Cov(R_{M}, R_{GE}) $$

Where $ w_{GE}$ the weight of an individual stock in the portfolio (in this example GE), and the other term is the covariance with the market return

-Keeping in mind that GE's contribution to the portfolio return is $ w_{GE}E(R_{GE})$, we can get the reward-to-risk ratio for investments in GE:

$$ \frac{\text{GE's contribution to risk premium}}{\text{GE's contribution to variance}} = \frac{w_{GE}E(R_{GE})}{w_{GE}Cov(R_{GE}, R_{M})} = \frac{E(R_{GE})}{Cov(R_{GE}, R_{M})} $$

-The reward-to-risk ratio for investment in the market portfolio is:

$$ \frac{\text{Market risk premium}}{\text{Market variance}} = \frac{E(R_{M})}{\sigma_M^2} $$

-Which is known as the **market price of risk** because it quantifies the extra return that investors demand to bear portfolio risk

  - notice that for the *components* of the efficient portfolio, risk is measured in terms of covariance (its contribution to portfolio variance) while for the efficient portfolio, variance is the right measure

  -keep in mind that the market price of risk is not the Sharpe Ratio where the denominator is standard deviation

-A basic principle of equilibrium is that all investments should offer the same reward-to-risk ratio

  - If the ratio were off, trading would equalize them

  - Therefore the reward-to-risk ratio of GE and the market portfolio should be equal:

  $$ \frac{E(R_{GE})}{Cov(R_{GE},R_{M})} = \frac{E(R_M)}{\sigma_M^2} $$

$$ \implies E(R_{GE}) = \frac{Cov(R_{GE},R_{M})}{\sigma_M^2}E(R_M)$$

Where the ratio $\frac{Cov(R_{GE},R_{M})}{\sigma_M^2}$ is whats called **beta** and is denoted by $\beta$, so that:

$$ E(r_{GE}) = r_f + \beta_{GE}[E(r_M) - r_f] $$

The is the **expected return-beta (or mean-beta) relationship**

-This relationship tells us that the total expected rate of return is the sum of the risk-free rate (compensation for waiting) plus a risk premium (compensation for "worrying") which is itself a product of the benchmark risk premium (the broad market portfolio) and the relative risk of the particular asset

-Notice that the risk premium does *not* depend on the total volatility of the investment

  - e.g. a firm developing a new drug may have extremely high variance but investors will no demand a higher return because the success of the firm is independent of macroeconomic risk

  - essentially CAPM predicts that systematic risk should by "priced" (meaning that it commands a risk premium), but firm-specific risk should not be priced by the market

### The Security Market Line

-Recall the Capital Allocation Line (CAL) plotted the risk premiums of *efficient portfolios* as a function of portfolio standard deviation (The Capital Market Line was simply this graph but of portfolios consisting of the market index and a risk-free asset)

  - standard deviation is the valid measure of risk for efficiently diversified portfolios

-In contrast, the Security Market Line (SML) graphs *individual asset* risk premiums as a function of *asset risk*

  - The relevant measure of risk for individual assets held as parts of a well-diversified portfolio is not the standard deviation or variance, but the contribution of the asset to the portfolio variance ($\beta$)

-The SML provides a benchmark for the evaluation of investment performance!!!

  - "fairly priced" assets plot exactly on the SML

  - all securities lie on the SML in market equilibrium 

  - We can use CAPM within the investment realm: if we evalute a stock (using different methods) and determine it has an expected return which lies above the SML, it is underpriced and a good buy

-The difference between a stock's expected return and its fair value from CAPM is called the **$\alpha$**

-You can say that security analysis is about uncovering securities with nonzero alphas!

-CAPM suggests that the starting point of portfolio management is a passive market index portfolio, then the manager would increase the weights of securities with positive alphas and decrease those with negative alphas

  - One strategy for adjusting the weights like this is the Treynor-Black model

-CAPM is also useful for capital budgeting decisions

  - For a firm considering a new project, the CAPM can provide the *required rate of return* that the project needs to yield, based on its beta, to be acceptable to investors

  - Managers can use the CAPM to obtain this cutoff internal rate of return (IRR) or "hurdle rate"

### The CAPM and the Single-Index Model

-Recall, that the index model predicts:

$$ E(R_{i}) = \alpha_{i} + \beta_{i}E(R_{M}) $$

While the CAPM predicts:

$$ E(R_{i}) = \beta_{i}E(R_{M}) $$

-Essentially, CAPM tells us that in market equilibrium, all alphas have been driven to zero

  - It also implies that index modelling should yield values of alpha which cluster around zero
